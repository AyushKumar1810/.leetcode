```python
# Web3 Trading Team - Data Science Assignment
# Trader Behavior vs Market Sentiment Analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("üöÄ Starting Trader Behavior & Market Sentiment Analysis")
print("=" * 60)

# =============================================================================
# 1. DATA LOADING AND INITIAL EXPLORATION
# =============================================================================

print("\nüìä STEP 1: Loading Datasets")
print("-" * 40)

# Load the datasets
# Note: You'll need to download the files from the Google Drive links and upload to Colab
try:
    # Load Historical Trader Data
    trader_data = pd.read_csv('/content/historical_trader_data.csv')
    print(f"‚úÖ Trader data loaded: {trader_data.shape[0]} rows, {trader_data.shape[1]} columns")

    # Load Fear & Greed Index Data
    sentiment_data = pd.read_csv('/content/fear_greed_index.csv')
    print(f"‚úÖ Sentiment data loaded: {sentiment_data.shape[0]} rows, {sentiment_data.shape[1]} columns")

except FileNotFoundError:
    print("‚ö†Ô∏è  Please upload the datasets to Colab first!")
    print("1. Download files from the provided Google Drive links")
    print("2. Upload to Colab using the file upload feature")
    print("3. Ensure filenames match: 'historical_trader_data.csv' and 'fear_greed_index.csv'")

# Display basic information about datasets
print("\nüîç Trader Data Overview:")
display(trader_data.head())
print(f"\nColumns: {list(trader_data.columns)}")
print(f"\nData types:\n{trader_data.dtypes}")

print("\nüîç Sentiment Data Overview:")
display(sentiment_data.head())
print(f"\nColumns: {list(sentiment_data.columns)}")
print(f"\nData types:\n{sentiment_data.dtypes}")

# =============================================================================
# 2. DATA PREPROCESSING AND CLEANING
# =============================================================================

print("\nüßπ STEP 2: Data Preprocessing")
print("-" * 40)

# Clean and preprocess trader data
def preprocess_trader_data(df):
    df_clean = df.copy()

    # Convert time column to datetime
    if 'Timestamp IST' in df_clean.columns:
        # Specify the format to avoid errors
        df_clean['time'] = pd.to_datetime(df_clean['Timestamp IST'], format='%d-%m-%Y %H:%M', errors='coerce')
        df_clean['date'] = df_clean['time'].dt.date

    # Handle missing values
    print(f"Missing values before cleaning:\n{df_clean.isnull().sum()}")

    # Convert numeric columns
    numeric_columns = ['Execution Price', 'Size Tokens', 'Size USD', 'Closed PnL', 'Fee', 'Timestamp']
    for col in numeric_columns:
        if col in df_clean.columns:
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

    # Rename columns for consistency
    df_clean.rename(columns={'Account': 'account', 'Coin': 'symbol', 'Execution Price': 'execution price',
                             'Size Tokens': 'size_tokens', 'Size USD': 'size_usd', 'Side': 'side',
                             'Start Position': 'start_position', 'Direction': 'direction', 'Closed PnL': 'closedPnL',
                             'Transaction Hash': 'transaction_hash', 'Order ID': 'order_id', 'Crossed': 'crossed',
                             'Fee': 'fee', 'Trade ID': 'trade_id', 'Timestamp': 'timestamp_unix'}, inplace=True)


    # Remove rows with critical missing data
    df_clean = df_clean.dropna(subset=['account'])

    print(f"‚úÖ Trader data cleaned: {df_clean.shape[0]} rows remaining")
    return df_clean

# Clean and preprocess sentiment data
def preprocess_sentiment_data(df):
    df_clean = df.copy()

    # Convert Date column to datetime
    if 'date' in df_clean.columns:
        # Assuming the 'date' column is already in a format that pandas can infer
        df_clean['Date'] = pd.to_datetime(df_clean['date'], errors='coerce')
        df_clean['date'] = df_clean['Date'].dt.date

    # Standardize classification values
    if 'classification' in df_clean.columns:
        df_clean['classification'] = df_clean['classification'].str.lower().str.strip()
        # Map more detailed classifications to 'fear' or 'greed'
        df_clean['classification'] = df_clean['classification'].replace({
            'extreme fear': 'fear',
            'extreme greed': 'greed',
            'neutral': 'neutral' # Keep neutral if present
        })
        df_clean['sentiment_numeric'] = df_clean['classification'].map({'fear': 0, 'neutral': 0.5, 'greed': 1}) # Add neutral mapping

    print(f"‚úÖ Sentiment data cleaned: {df_clean.shape[0]} rows")
    return df_clean

# Apply preprocessing
trader_clean = preprocess_trader_data(trader_data)
sentiment_clean = preprocess_sentiment_data(sentiment_data)

# Ensure date columns are datetime objects before merging
trader_clean['date'] = pd.to_datetime(trader_clean['date'], errors='coerce').dt.date
sentiment_clean['date'] = pd.to_datetime(sentiment_clean['date'], errors='coerce').dt.date


# Check date types and unique values before merging
print("\nChecking date types and unique values before merge:")
print(f"Trader data 'date' type: {trader_clean['date'].dtype}")
print(f"Sentiment data 'date' type: {sentiment_clean['date'].dtype}")
print(f"Unique dates in trader data: {trader_clean['date'].nunique()}")
print(f"Unique dates in sentiment data: {sentiment_clean['date'].nunique()}")
print(f"Sample dates in trader data: {trader_clean['date'].dropna().unique()[:5]}")
print(f"Sample dates in sentiment data: {sentiment_clean['date'].dropna().unique()[:5]}")


# =============================================================================
# 3. EXPLORATORY DATA ANALYSIS (EDA)
# =============================================================================

print("\nüìà STEP 3: Exploratory Data Analysis")
print("-" * 40)

# Create output directory for plots
import os
os.makedirs('/content/outputs', exist_ok=True)

# 3.1 Trader Performance Analysis
def analyze_trader_performance(df):
    print("\nüéØ Trader Performance Analysis")

    # Calculate key metrics
    trader_metrics = df.groupby('account').agg({
        'closedPnL': ['sum', 'mean', 'count'],
        'size_usd': ['sum', 'mean'], # Use size_usd for volume
    }).round(4)

    trader_metrics.columns = ['total_pnl', 'avg_pnl', 'trade_count', 'total_volume_usd', 'avg_size_usd']
    trader_metrics['profit_ratio'] = (trader_metrics['total_pnl'] > 0).astype(int)

    print(f"üìä Trader Performance Summary:")
    display(trader_metrics.describe())

    # Visualization: PnL Distribution
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.hist(trader_metrics['total_pnl'], bins=50, alpha=0.7, color='skyblue')
    plt.title('Distribution of Total PnL')
    plt.xlabel('Total PnL')
    plt.ylabel('Number of Traders')

    plt.subplot(1, 2, 2)
    plt.hist(trader_metrics['trade_count'], bins=30, alpha=0.7, color='lightgreen')
    plt.title('Distribution of Trade Count')
    plt.xlabel('Number of Trades')
    plt.ylabel('Number of Traders')

    plt.tight_layout()
    plt.savefig('/content/outputs/trader_performance_overview.png', dpi=300, bbox_inches='tight')
    plt.show()

    return trader_metrics

# 3.2 Market Sentiment Analysis
def analyze_market_sentiment(df):
    print("\nüìä Market Sentiment Analysis")

    # Sentiment distribution
    sentiment_counts = df['classification'].value_counts()
    print(f"Sentiment Distribution:\n{sentiment_counts}")

    # Visualization
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['red', 'grey', 'green']) # Added grey for neutral
    plt.title('Market Sentiment Distribution')
    plt.ylabel('')

    plt.subplot(1, 2, 2)
    df['classification'].value_counts().plot(kind='bar', color=['red', 'grey', 'green']) # Added grey for neutral
    plt.title('Fear vs Neutral vs Greed Count')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.xticks(rotation=0)

    plt.tight_layout()
    plt.savefig('/content/outputs/market_sentiment_overview.png', dpi=300, bbox_inches='tight')
    plt.show()

    return sentiment_counts

# Run EDA
trader_metrics = analyze_trader_performance(trader_clean)
sentiment_summary = analyze_market_sentiment(sentiment_clean)

# =============================================================================
# 4. CORRELATION ANALYSIS: TRADER BEHAVIOR vs MARKET SENTIMENT
# =============================================================================

print("\nüîó STEP 4: Correlation Analysis")
print("-" * 40)

# Merge datasets on date
def merge_datasets(trader_df, sentiment_df):
    # Aggregate trader data by date
    daily_trader_metrics = trader_df.groupby('date').agg({
        'closedPnL': ['sum', 'mean', 'count'],
        'size_usd': ['sum', 'mean'], # Use size_usd for volume
        'account': 'nunique'
    }).round(4)

    print("\nColumns after aggregation in merge_datasets:")
    print(daily_trader_metrics.columns)

    daily_trader_metrics.columns = ['total_daily_pnl', 'avg_daily_pnl', 'daily_trades',
                                   ('size_usd', 'sum'), ('size_usd', 'mean'), 'active_traders']


    daily_trader_metrics = daily_trader_metrics.reset_index()

    # Merge with sentiment data
    merged_data = pd.merge(daily_trader_metrics, sentiment_df, on='date', how='left')

    print(f"‚úÖ Merged dataset created: {merged_data.shape[0]} days of data")
    return merged_data

merged_analysis = merge_datasets(trader_clean, sentiment_clean)

# Analyze correlation between sentiment and trading behavior
def correlation_analysis(df):
    print("\nüîç Correlation Analysis Results")

    # Compare metrics by sentiment
    sentiment_comparison = df.groupby('classification').agg({
        'total_daily_pnl': ['mean', 'std', 'sum'],
        'avg_daily_pnl': ['mean', 'std'],
        'daily_trades': ['mean', 'std'],
        ('size_usd', 'sum'): ['mean', 'std'], # Use size_usd for volume
        'active_traders': ['mean', 'std']
    }).round(4)

    print("üìä Trading Metrics by Market Sentiment:")
    display(sentiment_comparison)

    # Statistical significance testing
    from scipy import stats

    fear_data = df[df['classification'] == 'fear']
    greed_data = df[df['classification'] == 'greed']
    neutral_data = df[df['classification'] == 'neutral'] # Added neutral data

    metrics_to_test = ['total_daily_pnl', 'avg_daily_pnl', 'daily_trades']

    print("\nüìà Statistical Significance Tests (t-test):")
    # Compare Fear vs Greed
    print("\nFear vs Greed:")
    for metric in metrics_to_test:
        if metric in df.columns:
            t_stat, p_value = stats.ttest_ind(fear_data[metric].dropna(), greed_data[metric].dropna())
            significance = "Significant" if p_value < 0.05 else "Not Significant"
            print(f"{metric}: t-stat={t_stat:.4f}, p-value={p_value:.4f} - {significance}")

    # Compare Fear vs Neutral
    if not neutral_data.empty:
        print("\nFear vs Neutral:")
        for metric in metrics_to_test:
            if metric in df.columns:
                t_stat, p_value = stats.ttest_ind(fear_data[metric].dropna(), neutral_data[metric].dropna())
                significance = "Significant" if p_value < 0.05 else "Not Significant"
                print(f"{metric}: t-stat={t_stat:.4f}, p-value={p_value:.4f} - {significance}")

        # Compare Greed vs Neutral
        print("\nGreed vs Neutral:")
        for metric in metrics_to_test:
            if metric in df.columns:
                t_stat, p_value = stats.ttest_ind(greed_data[metric].dropna(), neutral_data[metric].dropna())
                significance = "Significant" if p_value < 0.05 else "Not Significant"
                print(f"{metric}: t-stat={t_stat:.4f}, p-value={p_value:.4f} - {significance}")


    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # PnL comparison
    df.boxplot(column='total_daily_pnl', by='classification', ax=axes[0,0])
    axes[0,0].set_title('Daily PnL by Market Sentiment')
    axes[0,0].set_xlabel('Market Sentiment')

    # Trade volume comparison
    df.boxplot(column='daily_trades', by='classification', ax=axes[0,1])
    axes[0,1].set_title('Daily Trade Count by Market Sentiment')
    axes[0,1].set_xlabel('Market Sentiment')

    # Active traders comparison
    df.boxplot(column='active_traders', by='classification', ax=axes[1,1])
    axes[1,1].set_title('Active Traders by Market Sentiment')
    axes[1,1].set_xlabel('Market Sentiment')

    plt.tight_layout()
    plt.savefig('/content/outputs/sentiment_behavior_correlation.png', dpi=300, bbox_inches='tight')
    plt.show()

    return sentiment_comparison

correlation_results = correlation_analysis(merged_analysis)

# =============================================================================
# 5. ADVANCED PATTERN DISCOVERY
# =============================================================================

print("\nüî¨ STEP 5: Advanced Pattern Discovery")
print("-" * 40)

# 5.1 Sentiment Transition Analysis
def sentiment_transition_analysis(df):
    print("\nüîÑ Sentiment Transition Impact Analysis")

    # Create sentiment shifts
    df_sorted = df.sort_values('date').reset_index(drop=True)
    df_sorted['prev_sentiment'] = df_sorted['classification'].shift(1)
    df_sorted['sentiment_shift'] = df_sorted['prev_sentiment'] + ' -> ' + df_sorted['classification']

    # Analyze performance during transitions
    transition_analysis = df_sorted.groupby('sentiment_shift').agg({
        'total_daily_pnl': ['mean', 'sum', 'count'],
        'daily_trades': 'mean'
    }).round(4)

    print("üìä Performance During Sentiment Transitions:")
    display(transition_analysis)

    # Visualization
    plt.figure(figsize=(12, 6))

    transition_pnl = df_sorted.groupby('sentiment_shift')['total_daily_pnl'].mean()
    transition_pnl.plot(kind='bar', color=['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'brown', 'pink', 'gray']) # Added more colors
    plt.title('Average Daily PnL by Sentiment Transition')
    plt.ylabel('Average Daily PnL')
    plt.xticks(rotation=45, ha='right') # Rotate labels and align right


    plt.tight_layout()
    plt.savefig('/content/outputs/sentiment_transition_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

    return transition_analysis

# 5.2 Trader Profitability Segmentation
def trader_profitability_analysis(trader_df, sentiment_df):
    print("\nüí∞ Trader Profitability Segmentation")

    # Create trader performance segments
    trader_performance = trader_df.groupby('account').agg({
        'closedPnL': 'sum',
        'size_usd': 'sum', # Use size_usd for volume
    }).round(4)

    # Define profitability segments
    trader_performance['profitability_segment'] = pd.cut(
        trader_performance['closedPnL'],
        bins=[-np.inf, -1000, 0, 1000, np.inf],
        labels=['High Loss', 'Small Loss', 'Small Profit', 'High Profit']
    )

    # Merge with daily sentiment data
    trader_df_with_segment = trader_df.merge(
        trader_performance[['profitability_segment']],
        left_on='account', right_index=True
    )

    # Analyze behavior by segment and sentiment
    segment_sentiment_analysis = trader_df_with_segment.groupby(['date', 'profitability_segment']).agg({
        'size_usd': 'mean', # Use size_usd for volume
        'closedPnL': 'mean'
    }).reset_index()

    # Merge with sentiment
    segment_sentiment_analysis = segment_sentiment_analysis.merge(sentiment_df, on='date')

    # Compare segments across sentiments
    final_analysis = segment_sentiment_analysis.groupby(['profitability_segment', 'classification']).agg({
        'size_usd': 'mean', # Use size_usd for volume
        'closedPnL': 'mean'
    }).round(4)

    print("üìä Trading Behavior by Profitability Segment and Market Sentiment:")
    display(final_analysis)

    return final_analysis

# Run advanced analyses
if len(merged_analysis) > 1:
    transition_results = sentiment_transition_analysis(merged_analysis)
    profitability_results = trader_profitability_analysis(trader_clean, sentiment_clean)

# =============================================================================
# 6. KEY INSIGHTS AND RECOMMENDATIONS
# =============================================================================

print("\nüí° STEP 6: Key Insights and Strategic Recommendations")
print("=" * 60)

def generate_insights(correlation_results, transition_results=None):
    insights = []

    print("üéØ KEY FINDINGS:")
    print("-" * 20)

    # Insight 1: PnL differences between sentiments
    # Check if 'fear' and 'greed' exist in index
    if 'fear' in correlation_results.index and 'greed' in correlation_results.index:
        fear_pnl = correlation_results.loc['fear', ('total_daily_pnl', 'mean')]
        greed_pnl = correlation_results.loc['greed', ('total_daily_pnl', 'mean')]

        if fear_pnl > greed_pnl:
            # Avoid division by zero or very small numbers
            percentage_diff = ((fear_pnl - greed_pnl) / abs(greed_pnl)) * 100 if greed_pnl != 0 else float('inf')
            insight = f"üìà CONTRARIAN SIGNAL: Traders perform {percentage_diff:.1f}% better during FEAR periods"
            insights.append(insight)
            print(f"1. {insight}")
        else:
            percentage_diff = ((greed_pnl - fear_pnl) / abs(fear_pnl)) * 100 if fear_pnl != 0 else float('inf')
            insight = f"üìà MOMENTUM SIGNAL: Traders perform {percentage_diff:.1f}% better during GREED periods"
            insights.append(insight)
            print(f"1. {insight}")
    else:
        print("1. Insufficient data for Fear vs Greed PnL comparison.")


    # Insight 2: Leverage behavior
    # Removed leverage comparison as column is missing


    # Insight 3: Trading activity
    if 'fear' in correlation_results.index and 'greed' in correlation_results.index:
        fear_trades = correlation_results.loc['fear', ('daily_trades', 'mean')]
        greed_trades = correlation_results.loc['greed', ('daily_trades', 'mean')]

        if fear_trades > greed_trades:
            percentage_diff = ((fear_trades - greed_trades) / abs(greed_trades)) * 100 if greed_trades != 0 else float('inf')
            insight = f"üìä ACTIVITY SPIKE: {percentage_diff:.1f}% more trading activity during FEAR periods"
            insights.append(insight)
            print(f"2. {insight}") # Changed from 3 to 2 after removing previous insight
        else:
            percentage_diff = ((greed_trades - fear_trades) / abs(fear_trades)) * 100 if fear_trades != 0 else float('inf')
            insight = f"üìä GREED ACTIVITY: {percentage_diff:.1f}% more trading activity during GREED periods"
            insights.append(insight)
            print(f"2. {insight}") # Changed from 3 to 2 after removing previous insight
    else:
         print("2. Insufficient data for Fear vs Greed trading activity comparison.") # Changed from 3 to 2

    print("\nüöÄ STRATEGIC RECOMMENDATIONS:")
    print("-" * 30)

    recommendations = [
        "üéØ Implement contrarian strategies during extreme sentiment periods",
        "‚öñÔ∏è Adjust position sizing based on market sentiment cycles",
        "üìà Monitor sentiment transitions for optimal entry/exit timing",
        "üîÑ Develop sentiment-aware risk management protocols",
        "üìä Create composite signals combining sentiment and trader behavior"
    ]

    for i, rec in enumerate(recommendations, 1):
        print(f"{i}. {rec}")

    return insights, recommendations

insights, recommendations = generate_insights(correlation_results,
                                            transition_results if 'transition_results' in locals() else None)

# =============================================================================
# 7. SAVE PROCESSED DATA AND RESULTS
# =============================================================================

print("\nüíæ STEP 7: Saving Results")
print("-" * 40)

# Create CSV files directory
os.makedirs('/content/csv_files', exist_ok=True)

# Save processed datasets
merged_analysis.to_csv('/content/csv_files/merged_trader_sentiment_data.csv', index=False)
trader_metrics.to_csv('/content/csv_files/trader_performance_metrics.csv', index=True)
correlation_results.to_csv('/content/csv_files/sentiment_correlation_analysis.csv', index=True)

# Save insights as text file
with open('/content/csv_files/key_insights.txt', 'w') as f:
    f.write("WEB3 TRADING TEAM - DATA SCIENCE ANALYSIS RESULTS\n")
    f.write("=" * 50 + "\n\n")
    f.write("KEY INSIGHTS:\n")
    for i, insight in enumerate(insights, 1):
        f.write(f"{i}. {insight}\n")
    f.write("\nSTRATEGIC RECOMMENDATIONS:\n")
    for i, rec in enumerate(recommendations, 1):
        f.write(f"{i}. {rec}\n")

print("‚úÖ All results saved successfully!")
print("\nFiles created:")
print("üìÅ /content/csv_files/merged_trader_sentiment_data.csv")
print("üìÅ /content/csv_files/trader_performance_metrics.csv")
print("üìÅ /content/csv_files/sentiment_correlation_analysis.csv")
print("üìÅ /content/csv_files/key_insights.txt")
print("üìÅ /content/outputs/ (contains all visualization PNG files)")

print("\nüéâ ANALYSIS COMPLETE!")
print("=" * 60)
print("Your comprehensive trader behavior vs market sentiment analysis is ready!")
print("All outputs follow the required submission structure.")
```
